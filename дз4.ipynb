```python
import matplotlib.pyplot as plt
import random
import math
plt.style.use('seaborn-whitegrid')

X = [ 15, 20, 23, 25 ]
Y = [  5,  10, 20, 60 ]


w= 0 
b = 0
#u = 0 
def neuronFire(x):
    global w,b
    y = w*x*x  + b
    return y

def batchProcess(X):
    Ypred = []
    for i in range(len(X)):
        x = X[i]
        y = neuronFire(x)
        Ypred.append(y)
    return Ypred  

def meanError(Yreal, Ypredicted):
    E = []
    for i in range(len(Yreal)):
        E.append(abs(Yreal[i] - Ypredicted[i]))
    
    e = sum(E) / len(E)
    return e

num_epochs = 150
max_error = 3

Errors= [100]
Weights = [w]
Biases =  [b]
Uises  = [u]
for epoch in range(num_epochs):
    
    print(f"\n###### START epoch: {epoch} #######")
    ##################epoch start #################
    dW = random.uniform(-5,5)
    w += dW
    Ypred = batchProcess(X)
    e = meanError(Y,Ypred)
    if epoch > 0:
        if e > Errors[epoch -1]:
            w -= dW
    
    dB = random.uniform(-5,5)
    b += dB
    Ypred = batchProcess(X)
    e = meanError(Y,Ypred)
    if epoch > 0:
        if e > Errors[epoch -1]:
            b -= dB
            
    #dU = random.uniform(-5,5)
    #u += dU
    #Ypred = batchProcess(X)
    #e = meanError(Y,Ypred)
    #if epoch > 0:
        #if e > Errors[epoch -1]:
            #u -= dU
    
    Errors.append(e)
    Weights.append(w)
    Biases.append(b)
    Uises.append(u)

    if e < max_error:
        break
    
    
    print("Yreal", Y)
    print("Ypred", Ypred)
    print(F"Mean error: {e:5.2f} %")
    print(f"####### END epoch: {epoch} ######\n")
    ################ epoch end ################## 

print(f"min error: {min(Errors)}") 

plt.plot([x for x  in range (epoch+2)], Errors, color='red', linestyle='solid', linewidth = 1, marker = "x")
plt.plot([x for x  in range (epoch+2)], Weights, color='black', linestyle='solid', linewidth = 1, marker = ".")
plt.plot([x for x  in range (epoch+2)], Biases, color='blue', linestyle='solid', linewidth = 1, marker = "+")
plt.plot([x for x  in range (epoch+2)], Uises, color='blue', linestyle='solid', linewidth = 1, marker = "+")
            
```
